## 문제

### score_grade에서 if문을 사용해 95점 이상일 때는, "Goob Job"도 함께 출력해주세요.


입력 : 98<br>
출력 : A<br>
출력 : Good Job<br>

```python
score = int(input('입력 : '))

if score >= 90:
    print('출력 : A')
    if score >= 95:
        print('출력 : Good Job')
elif 80 <= score < 90:
    print('출력 : B')
elif 70 <= score < 80:
    print('출력 : C')
elif 60 <= score < 70:
    print('출력 : D')
else:
    print('출력 : F')
```
