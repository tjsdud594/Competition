## 문제

### 파이썬의 if elif else 제어문을 활용하여 다음에 score에 따라서 입력된 점수의 등급을 출력하시오.

90점 이상 : A<br>
80점 이상 : B<br>
70점 이상 : C<br>
60점 이상 : D<br>
60점 미만 : F<br>

입력 : 81<br>
출력 : B

```python
score = int(input('입력 : '))

if score >= 90:
    print('출력 : A')
elif 80 <= score < 90:
    print('출력 : B')
elif 70 <= score < 80:
    print('출력 : C')
elif 60 <= score < 70:
    print('출력 : D')
else:
    print('출력 : F')
```
